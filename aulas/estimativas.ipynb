Variáveis preditoras: as com maior correlação com o alvo

```python
import statsmodels.api as sm

y = df['log_target_normalizado']

features = [
    'log_target',
    'valorPagoMaior1000',
    'valorPagoEmpenho',
    'te_uniEle_mean',
    'te_orgProgEle_mean',
    'te_progEle_mean',
    'te_uniEle_median',
    'te_orgProgEle_median',
    'te_progEle_median',
    'te_orgEle_mean',
    'te_funEle_mean',
    'te_subEle_mean',
    'te_orgEle_median',
    'te_funEle_median',
    'te_subEle_median',
    'te_descricaoElemento_mean',
    'te_descricaoElemento_median',
    'te_despesaDot_mean'
]

X = df[features]
X = sm.add_constant(X)

modelo_ols = sm.OLS(y, X).fit()
print(modelo_ols.summary())
```

```python
# Predict de uma observação existente no df para validação
idx = 10  # índice da linha escolhida para validar

X_novo = df.loc[[idx], features]
X_novo = sm.add_constant(X_novo, has_constant='add')

y_pred = modelo_ols.predict(X_novo)
y_real = df.loc[idx, 'log_target_normalizado']

print(f'Índice usado: {idx}')
print(f'Valor real  (log_target_normalizado): {y_real:.4f}')
print(f'Valor predito (OLS):                  {y_pred.values[0]:.4f}')
print(f'Diferença absoluta:                   {abs(y_real - y_pred.values[0]):.4f}')
```
